```python
from pathlib import Path

candidate_paths = [
	#Path("data/twitter_sentiment.csv"),
	Path("../data/twitter_sentiment.csv")
]

csv_path = next((p for p in candidate_paths if p.exists()), None)

if csv_path is None:
	print("Dataset not found. Checked:")
	for p in candidate_paths:
		print(f"- {p.resolve()}")
	df = pd.DataFrame()
else:
	df = pd.read_csv(csv_path)
	print(f"Data loaded successfully from: {csv_path.resolve()}")
```

```python
# Convert the column to object type to allow string values
df['category'] = df['category'].astype('object')

# Convert numeric categories to string labels
df.loc[df['category'] == 1.0, 'category'] = "positive"
df.loc[df['category'] == 0.0, 'category'] = "neutral"
df.loc[df['category'] == -1.0, 'category'] = "negative"
```
